```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display

# grid
n = 40
dx = 1.0

# distance field
z = np.full((n,n), np.inf)

# source point
cx = n//2
cy = n//2
z[cx,cy] = 0

# figure
fig, ax = plt.subplots(figsize=(8,8))
im = ax.imshow(z, cmap="magma", vmin=0, vmax=50, interpolation="nearest")

ax.set_title("Eikonal Wave Propagation")

def update(frame):
  global z

  z_new = z.copy()

  for i in range(1,n-1):
    for j in range(1,n-1):

      neighbors = [
        z[i-1,j],
        z[i+1,j],
        z[i,j-1],
        z[i,j+1]
      ]

      z_new[i,j] = min(z[i,j], min(neighbors)+dx)

  z[:] = z_new
  im.set_array(z)

  return [im]

anim = FuncAnimation(fig, update, frames=80, interval=50)

plt.close(fig)

display(HTML(anim.to_jshtml()))
```
